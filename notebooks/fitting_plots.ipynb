## Line on 2 points

```python
# Define the two points (X, Y) that the line passes through
X = np.array([1, 2])
Y = np.array([3, 5])
```

```python
# Fit the line using linear regression (1st-degree polynomial fit)
coefficients = np.polyfit(X, Y, 1)
intercept, slope = coefficients
```

```python
# Create a range of X values for plotting the line
x_range = np.linspace(min(X), max(X), 100)
y_range = intercept + slope * x_range

# Create the plot
plt.figure()
plt.plot(X, Y, 'bo', label='Data points')  # Plot the data points
plt.plot(x_range, y_range, 'r-', label='Line')  # Plot the line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Passing Through Two Points')
plt.legend()

# Show the plot
plt.show()
```

## Plot linear regression on 3D

```python
# Define the data points (X, Y, Z)
X = np.array([1, 2, 3])
Y = np.array([2, 3, 4])
Z = np.array([3, 5, 7])
```

```python
# Create the design matrix for the hyperplane
A = np.column_stack((X, Y, np.ones_like(X)))

# Calculate the coefficients of the hyperplane
coefficients, _, _, _ = np.linalg.lstsq(A, Z, rcond=None)
intercept, slope_x, slope_y = coefficients
```

```python
# Create a meshgrid for the plane
xx, yy = np.meshgrid(np.linspace(min(X), max(X), 100), np.linspace(min(Y), max(Y), 100))
zz = intercept + slope_x * xx + slope_y * yy

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(X, Y, Z, color='blue', label='Data points')

# Plot the hyperplane
ax.plot_surface(xx, yy, zz, alpha=0.5, color='red', label='Hyperplane')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Hyperplane through Three Points')

# Add a legend
# ax.legend()

# Show the plot
plt.show()
```
